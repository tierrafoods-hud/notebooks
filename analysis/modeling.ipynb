```python
import pandas as pd

country_name = 'Guatemala'
dataset_path = f'outputs/20241127/{country_name}_wosis/{country_name}_wosis_merged.csv'

# load datasets
dataset = pd.read_csv(dataset_path)

# check dataset
dataset.head()
```

```python
import matplotlib.pyplot as plt

# plot histogram of values
dataset.hist(bins=50, figsize=(20, 15))
plt.title(f'{country_name} WOSIS Merged Dataset')
plt.show()
```

```python
import seaborn as sns

# plot missing values
sns.heatmap(dataset.isnull(), cbar=False, cmap='viridis')
plt.title(f'{country_name} WOSIS Merged Dataset Missing Values')
plt.show()
```
